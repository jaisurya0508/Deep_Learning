# <center> 2.Splitting and Merging image channels using OpenCV

### Import libraries

```python
import cv2
import glob
import os
import matplotlib.pyplot as plt
```

### Read images

```python
img=cv2.imread('num.png')
plt.imshow(img)
```

### split this image into individual channels

```python
b,g,r=cv2.split(img)
```

```python
cv2.imshow('Blue',b)
cv2.waitKey(0)
cv2.imshow('Green',g)
cv2.waitKey(0)
cv2.imshow('Red',r)
cv2.waitKey(0)
```

### merge individual color channels to form a color image

```python
image=cv2.merge((b,g,r))
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
