# <center>1.Color Models and convert One color models to another color models

### Read image

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("ball.jpg")
```

### Convert Image to RGB color:

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Convert Images to HSV color coding

```python
img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

### Convert image to HLS color coding

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
